```
#ボード線図作成
from control.matlab import *
import matplotlib.pyplot as plt
import numpy as np

#伝達関数のパラメータ
#num = [2, 5, 1]
#den = [1, 2, 3]
sys = ss([[0, 1], [-5, -6]],[[0],[1]],[1, 0],0)
#sys = tf(num, den)
bode(sys)
plt.show()


#ステップ応答
t = np.linspace(0,5,1000)
y, T = step(sys,t)
plt.plot(T,y)
plt.grid()
plt.axhline(1, color="b", linestyle="--")
plt.xlim(0,3)

plt.show()
```
